```python
def series(x, scale=0.85, inc=0.1):
    """ Decorator to build a series given a value
    
    Parameters
    ----------
    x : str
        argument for x axis to be evaluated. Supports
            s, k, r, q, t, vol
    scale : float
        starts at passed s, k, r, q, t, vol and starts
            and ends the range of values between scale
    inc : float
        increment for the x axis argument
    
    """
    def _expand(init, scale, inc):
        start = init * scale
        stop = init * (1.0 + (1.0 - scale))
        data = []
        for i in xrange(len(start)):
            data.append(np.arange(start[i], stop[i] + inc, inc))
        return np.array(data)

    def _series(func):
        @wraps(func)
        def _build_series(user_func, **kwargs):
            init = np.ones(np.shape(kwargs[x])) * kwargs[x].mean()
            kwargs[x] = _expand(init, scale, inc)
            kwargs['index'] = _expand(init, scale, inc)
            return func(user_func, **kwargs)        
        return _build_series
    return _series
```

```python
#@series(x='s', scale=0.85, inc=0.01)
def build(user_func, **kwargs):
    """
    Evaluates user_func across a range of values
    
    Parameters
    ----------
    user_func : str
        the name of the function. Supports
            payoff, delta, gamma, vega, theta, rho
    args: dictionary
        the full set of parameters of the position
        logic is built in to use only the required
        functions
        
    Examples
    ----------
    # using s in the decorator
    
    # ratio call spread
    multi_leg = {
        's': np.array([126.75, 126.75]),
        'k': np.array([125.0, 127.50]),
        'r': np.array([0.0025, 0.0025]),
        'q': np.array([0.0, 0.0]),
        't': np.array([0.0/365.0, 0.0/365.0]),
        'vol': np.array([0.2791, 0.2791]),
        'option_type': np.array(['call', 'call']),
        'option_premium': np.array([4.65, 2.70]),
        'option_trade_qty': np.array([1.0, -2.0]),
        'underlying_price': np.array([126.75, 126.75]),
        'underlying_trade_qty': np.array([0.0, 0.0]),
        'underlying_shares': np.array([100.0, 100.0])
    }
    
    # payoff of a multi leg position across strikes
    result = evaluate('payoff', **multi_leg)
    
    # delta of a multi leg position across strikes
    result = evaluate('delta', **multi_leg)
    
    # gamma of a multi leg position across strikes
    result = evaluate('gamma', **multi_leg)
    
    # using t in the decorator
    
    # theta of a multi leg position across time
    result = evaluate('theta', **multi_leg)
    
    """
    
    position = []
    # for each option type passed
    for i in xrange(len(kwargs['option_type'])):
        
        # parse and get the function to be evaluated
        meth_name = '{0}_{1}'.format(kwargs['option_type'][i], user_func)
        func = getattr(bs, meth_name)
        
        # get the arg_spec collection
        arg_spec = inspect.getargspec(getattr(bs, meth_name))
        
        # get the arguments to be passed to the function
        func_args = (kwargs[key][i] for key in arg_spec[0])
        
        # map the arguments to the function and append array to position
        position.append(list(itertools.starmap(func, [func_args])))

    # sum the values between the positions
    return {
        'position': [sum(i) for i in zip(*position)][0],
        'components': np.hstack([i[0].reshape(-1, 1) for i in position]),
        #'index': kwargs['index'][0]
    }
```

```python
def build_position(options_spec, trade_spec):
    """
    
    """
    return {
        's': np.array(options_spec['s']),
        'k': np.array(options_spec['k']),
        'r': np.ones(np.shape(options_spec['s'])) * options_spec['r'],
        'q': np.ones(np.shape(options_spec['s'])) * options_spec['q'],
        't': np.ones(np.shape(options_spec['s'])) * options_spec['t'] / 365.0,
        'vol': np.array(options_spec['vol']),
        'option_type': options_spec['option_type'],
        'option_premium': np.array(options_spec['option_premium']),
        'option_trade_qty': np.array(options_spec['option_trade_qty']),
        'underlying_price': np.array(options_spec['s']),
        'underlying_trade_qty': np.ones(np.shape(options_spec['s'])) * trade_spec['underlying_trade_qty'],
        'underlying_shares': np.ones(np.shape(options_spec['s'])) * trade_spec['underlying_shares'],
    }

def plt(result, title):
    x = result['index']
    y0 = result['position']
    y1 = result['components']
    pandas.DataFrame(y0, x).plot(legend=False)
```

```python
def breakeven(position):

    x0 = position['underlying_price']
   
    if len(position['underlying_price']) > 1:
        position_last = position['option_premium'].sum()
        x0 = np.array([-1,1]) * (position_last + position['underlying_price'])

    def fcn(x0):
        position['s'] = x0
        result = build('payoff', **position)
        return result['position']

    x = fsolve(fcn, x0=x0, xtol=0.01, maxfev=500)
    dollar_change = x - position['underlying_price']
    percent_change = dollar_change / position['underlying_price']

    return x, dollar_change, percent_change
```
