```python
%matplotlib inline

import numpy as np
import pandas

import models
import models.bs as bs
```

```python
def plt(x, y, cols):
    pandas.DataFrame(y, index=x, columns=cols).plot()
```

```python
# option setup
s = np.arange(20.0, 40.0, 0.01)
k = np.array([30.0])
r = np.array([0.0025])
q = 0.0
t = np.array([0.0/365.0])
vol = np.array([0.2791])

option_type = np.array(['call'])
option_premium = np.array([1.5])
option_trade_qty = np.array([1.0])

underlying_price = np.array([35.0])
underlying_trade_qty = np.array([0.0])

payoff = bs.position_payoff(s, k, r, q, t, vol, option_type, option_premium, option_trade_qty,
                  underlying_price, underlying_trade_qty)

plt(s, payoff['position'], ['long call'])
```
